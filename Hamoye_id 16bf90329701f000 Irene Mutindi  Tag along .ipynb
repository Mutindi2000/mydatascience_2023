```python
#data visualization and presentation,Area,Item,Element and Unit
#!/usr/bin/env python
# coding: utf-8


import pandas as pd
import seaborn as sns
import warnings
from matplotlib import pyplot as plt
import webbrowser

#importing data and cleaning it
def dataimportation():
    data =pd.read_csv("C:\Hamoye\Project\data.csv", encoding="latin-1")
    data=data.drop(["Area Code", "Item Code", "Element Code"], axis =1)
    return data
    

#Clustered Bar Chart for Area and years from 2014 to 2018
    
def AreaBar ():
    warnings.filterwarnings("ignore")
    areadata=dataimportation()
    dfArea =areadata[["Area", "Y2014", "Y2015", "Y2016", "Y2017", "Y2018"]]
    data2014=dfArea[["Area", "Y2014"]]
    data2014=data2014.rename(columns={"Y2014":"Yearvalue"})
    data2014["Year"]="Year2014"
    data2015 =dfArea[["Area", "Y2015"]]
    data2015["Year"]="Year2015"
    data2015=data2015.rename(columns={"Y2015":"Yearvalue"})
    data2016 =dfArea[["Area", "Y2016"]]
    data2016["Year"]="Year2016"
    data2016=data2016.rename(columns={"Y2016":"Yearvalue"})
    data2017 =dfArea[["Area", "Y2017"]]
    data2017["Year"]="Year2017"
    data2017=data2017.rename(columns={"Y2017":"Yearvalue"})
    data2018 =dfArea[["Area", "Y2018"]]
    data2018["Year"]="Year2018"
    data2018=data2018.rename(columns={"Y2018":"Yearvalue"})
    Areadata= pd.concat([data2014, data2015, data2016, data2017, data2018], ignore_index=True)
    Areadata=Areadata.groupby(["Area", "Year"]).agg(mean =("Yearvalue", "mean"))
    Areadata=Areadata.reset_index()
    sns.set(font_scale=1.5)
    plt.figure(figsize=(25, 15))
    plt.title("BarPlot for Area(Country) Factored by the mean of the years From 2014 to 2018", fontsize =30, color ="blue")
    graph=sns.barplot(x="Area", y ="mean", hue="Year", data=Areadata)
    graph.set_xticklabels(graph.get_xticklabels(), rotation=90)
    plt.xlabel("Area[country]", fontsize =20)
    plt.ylabel ("Average [2014 to 2018]", fontsize =20)
    
    #Clustered barchart for item and years from 2014 to 2018
def BarItem ():
    warnings.filterwarnings("ignore")
    itemdata=dataimportation()
    dfitem =itemdata[["Item", "Y2014", "Y2015", "Y2016", "Y2017", "Y2018"]]
    data2014=dfitem[["Item", "Y2014"]]
    data2014=data2014.rename(columns={"Y2014":"Yearvalue"})
    data2014["Year"]="Year2014"
    data2015 =dfitem[["Item", "Y2015"]]
    data2015["Year"]="Year2015"
    data2015=data2015.rename(columns={"Y2015":"Yearvalue"})
    data2016 =dfitem[["Item", "Y2016"]]
    data2016["Year"]="Year2016"
    data2016=data2016.rename(columns={"Y2016":"Yearvalue"})
    data2017 =dfitem[["Item", "Y2017"]]
    data2017["Year"]="Year2017"
    data2017=data2017.rename(columns={"Y2017":"Yearvalue"})
    data2018 =dfitem[["Item", "Y2018"]]
    data2018["Year"]="Year2018"
    data2018=data2018.rename(columns={"Y2018":"Yearvalue"})
    itemdata= pd.concat([data2014, data2015, data2016, data2017, data2018], ignore_index=True)
    itemdata=itemdata[~itemdata["Yearvalue"].isnull()]
    itemdata=itemdata.groupby(["Item", "Year"]).agg(mean =("Yearvalue", "mean"))
    itemdata=itemdata.reset_index()
    sns.set(font_scale=2)
    plt.figure(figsize=(30, 20))
    plt.title("Clustered BarPlot for Item and Average from 2014 to 2018", fontsize =40, color ="blue")
    graph=sns.barplot(x="Item", y ="mean", hue="Year", data=itemdata)
    graph.set_xticklabels(graph.get_xticklabels(), rotation=90)
    plt.xlabel("Item", fontsize =40)
    plt.ylabel("Average [2014 to 2018]", fontsize =40)
    plt.xlim(-0.5, 50)
    
    
#Clustered Barplot for Elements Factored by Year averages from 2014 to 2018

def BarElement ():
    warnings.filterwarnings("ignore")
    Elementdata=dataimportation()
    dfElement =Elementdata[["Element", "Y2014", "Y2015", "Y2016", "Y2017", "Y2018"]]
    data2014=dfElement[["Element", "Y2014"]]
    data2014=data2014.rename(columns={"Y2014":"Yearvalue"})
    data2014["Year"]="Year2014"
    data2015 =dfElement[["Element", "Y2015"]]
    data2015["Year"]="Year2015"
    data2015=data2015.rename(columns={"Y2015":"Yearvalue"})
    data2016 =dfElement[["Element", "Y2016"]]
    data2016["Year"]="Year2016"
    data2016=data2016.rename(columns={"Y2016":"Yearvalue"})
    data2017 =dfElement[["Element", "Y2017"]]
    data2017["Year"]="Year2017"
    data2017=data2017.rename(columns={"Y2017":"Yearvalue"})
    data2018 =dfElement[["Element", "Y2018"]]
    data2018["Year"]="Year2018"
    data2018=data2018.rename(columns={"Y2018":"Yearvalue"})
    dfElement= pd.concat([data2014, data2015, data2016, data2017, data2018], ignore_index=True)
    dfElement=dfElement[~dfElement["Yearvalue"].isnull()]
    dfElement=dfElement.groupby(["Element", "Year"]).agg(mean =("Yearvalue", "mean"))
    dfElement=dfElement.reset_index()
    sns.set(font_scale=2)
    plt.figure(figsize=(16, 20))
    plt.title("Clustered BarPlot for Element and Average from 2014 to 2018", fontsize =20, color ="blue")
    graph=sns.barplot(x="Element", y ="mean", hue="Year", data=dfElement)
    graph.set_xticklabels(graph.get_xticklabels(), rotation=80)
    plt.xlabel("Element", fontsize =40)
    plt.ylabel("Average [2014 to 2018]", fontsize =40)
    
#Clustered BarPlot for Unit Factored by Year averages from 2014 to 2018

def BarUnit ():
    warnings.filterwarnings("ignore")
    Unitdata=dataimportation()
    Unitdata =Unitdata[["Unit", "Y2014", "Y2015", "Y2016", "Y2017", "Y2018"]]
    data2014=Unitdata[["Unit", "Y2014"]]
    data2014=data2014.rename(columns={"Y2014":"Yearvalue"})
    data2014["Year"]="Year2014"
    data2015 =Unitdata[["Unit", "Y2015"]]
    data2015["Year"]="Year2015"
    data2015=data2015.rename(columns={"Y2015":"Yearvalue"})
    data2016 =Unitdata[["Unit", "Y2016"]]
    data2016["Year"]="Year2016"
    data2016=data2016.rename(columns={"Y2016":"Yearvalue"})
    data2017 =Unitdata[["Unit", "Y2017"]]
    data2017["Year"]="Year2017"
    data2017=data2017.rename(columns={"Y2017":"Yearvalue"})
    data2018 =Unitdata[["Unit", "Y2018"]]
    data2018["Year"]="Year2018"
    data2018=data2018.rename(columns={"Y2018":"Yearvalue"})
    Unitdata= pd.concat([data2014, data2015, data2016, data2017, data2018], ignore_index=True)
    Unitdata=Unitdata[~Unitdata["Yearvalue"].isnull()]
    Unitdata=Unitdata.groupby(["Unit", "Year"]).agg(mean =("Yearvalue", "mean"))
    Unitdata=Unitdata.reset_index()
    sns.set(font_scale=2)
    plt.figure(figsize=(20, 20))
    plt.title("Clustered BarPlot for Unit and Average from 2014 to 2018", fontsize =20, color ="blue")
    graph=sns.barplot(x="Unit", y ="mean", hue="Year", data=Unitdata)
    graph.set_xticklabels(graph.get_xticklabels(), rotation=30)
    plt.xlabel("Unit", fontsize =40)
    plt.ylabel("Average [2014 to 2018]", fontsize =40)
    
    
  
AreaBar()
BarItem()
BarElement()
BarUnit()
```
